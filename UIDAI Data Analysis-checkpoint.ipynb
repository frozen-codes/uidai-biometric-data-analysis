```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os

# Folder where CSV files are stored
folder_path =r"C:\Users\suraj\Downloads\aadhar biometric data"

csv_files = glob.glob(os.path.join(folder_path, "api_data_aadhar_biometric_*.csv"))

df_list = [pd.read_csv(file) for file in csv_files]
df = pd.concat(df_list, ignore_index=True)

# Preprocessing
df['date'] = pd.to_datetime(df['date'], format="%d-%m-%Y")
df['total_biometric'] = df['bio_age_5_17'] + df['bio_age_17_']

print("Total Records:", len(df))
```

```python
df.info()
```

```python
plt.figure()
df.groupby('state')['total_biometric'].sum() \
  .sort_values(ascending=False).head(10) \
  .plot(kind='bar')

plt.title("Top 10 States by Aadhaar Biometric Usage")
plt.xlabel("State")
plt.ylabel("Total Authentications")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Visualization 1: Top 10 States by Aadhaar Biometric Usage
What the Visualization Shows

Maharashtra and Uttar Pradesh lead Aadhaar biometric usage by a large margin.

A steep drop is visible after the top 23 states.

Usage is highly concentrated, not evenly distributed.

Insight Extracted

Aadhaar biometric demand is population-driven but also service-access driven.

High-load states are more prone to:

Authentication failures

Server latency

Overcrowded enrollment/update centers

Solution Implication

 Targeted Infrastructure Scaling

Allocate more biometric devices and operators in high-usage states

Introduce load-based center planning instead of uniform allocation

You can phrase this as:

A data-driven infrastructure allocation model can significantly improve Aadhaar service efficiency by prioritizing high-demand states.

```python
daily_trend = df.groupby('date')['total_biometric'].sum()

plt.figure()
plt.plot(daily_trend.index, daily_trend.values)
plt.title("Monthly Aadhaar Biometric Authentication Trend")
plt.xlabel("Month")
plt.ylabel("Total Authentications")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
df.groupby('district')['total_biometric'].sum() \
  .sort_values(ascending=False).head(10) \
  .plot(kind='bar')

plt.title("Top 10 Districts by Biometric Usage")
plt.xlabel("District")
plt.ylabel("Total Authentications")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

 Key line for PDF:

District-level analysis reveals strong urban concentration, indicating the need for city-specific Aadhaar service optimization strategies.

```python
plt.figure()
df.groupby('state')['total_biometric'].mean() \
  .sort_values(ascending=False).head(10) \
  .plot(kind='bar')

plt.title("Top 10 States by Average Daily Biometric Usage")
plt.xlabel("State")
plt.ylabel("Average Daily Authentications")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

 Visualization 5: Top 10 States by Average Daily Biometric Usage
What the Visualization Shows

Chhattisgarh and Delhi rank high in average daily usage despite not being the highest in total volume.

Smaller or administratively efficient states show consistent daily demand.

High population states do not always top this list.

Insight Extracted

Total usage  daily operational pressure

States with moderate population but high consistency face continuous system load

Solution Implication

 Operational Efficiency Planning

Daily-average metrics should guide:

Staff scheduling

Server uptime planning

Uniform infrastructure allocation may cause inefficiencies

 Key line:

Average daily usage highlights operational stress points that total volume metrics alone fail to capture.

```python
rolling_trend = df.groupby('date')['total_biometric'].sum().rolling(7).mean()

plt.figure()
plt.plot(rolling_trend.index, rolling_trend.values)
plt.title("7-Day Rolling Average of Aadhaar Biometric Usage")
plt.xlabel("Date")
plt.ylabel("Average Authentications")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 Visualization 6: 7-Day Rolling Average of Aadhaar Biometric Usage
What the Visualization Shows

Rolling average smooths noise and reveals:

A clear rising trend till early December

A gradual decline afterward

Confirms the presence of cyclical demand patterns

Insight Extracted

Aadhaar usage follows temporal cycles

Likely linked to:

Welfare distribution windows

Academic schedules

Administrative deadlines

Solution Implication

 Predictive & Preventive Governance

Rolling trends can be used for:

Early warning systems

Dynamic resource allocation

Prevents system overload during predictable spikes

 Key line:

Rolling average analysis enables early detection of demand surges, supporting proactive Aadhaar system management.
